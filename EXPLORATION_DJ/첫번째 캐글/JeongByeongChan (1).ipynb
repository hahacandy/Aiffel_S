# 3. 모델 설계

```python
# 필요한 모델을 불러오겠습니다.
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
import lightgbm as lgb
```

```python
xgboost = xgb.XGBRegressor(random_state=2019)
lightgbm = lgb.LGBMRegressor(random_state=2019)
```

```python
# 그리드 서치에 필요한 파라미터 설정입니다.
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [1, 10],
}
```

```python
# 모델은 LGBMRegressor을 사용하겠습니다.
model = lgb.LGBMRegressor(random_state=2019)
```

```python
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):
    # GridSearchCV 모델로 초기화
    grid_model = GridSearchCV(model, param_grid=param_grid, scoring='neg_mean_squared_error', \
                              cv=5, verbose=verbose, n_jobs=n_jobs)

    # 모델 fitting
    grid_model.fit(train, y)

    # 결과값 저장
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']

    # 데이터 프레임 생성
    results = pd.DataFrame(params)
    results['score'] = score

    # RMSLE 값 계산 후 정렬
    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results = results.sort_values('RMSLE')

    return results
```

```python
# 그리드 서치 함수를 이용해 좋은 하이퍼 파라미터 수치를 찾아봅니다.
my_GridSearch(model, data, y, param_grid)
```

```python
def m_pre(model, x, y, sub_x):
    # 모델을 훈련 합니다.
    model.fit(x.values, y)
    # 훈련된 데이터를 토대로 예측값을 반환합니다.
    return model.predict(sub_x.values)

print('얍💢')
```

```python
# 위에서 찾은 하이퍼 파라미터로 모델을 생성합니다.
random_state = 100 
model = lgb.LGBMRegressor(max_depth=10, n_estimators=50, random_state=random_state)
```

```python
# 만들어진 모델로 예측값을 구해봅니다.
y_pred = m_pre(model, data, y, sub)
```
